# Weather Data Analysis

**Exploratory Data Analysis (EDA) pipeline using Python, Pandas, NumPy, and Matplotlib.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 1. Import and Load Dataset

```python
df = pd.read_csv('weather_data.csv')
df
```

## 2. Dataset Information

```python
df.info()
```

```python
df.shape
```

## 3. Data Cleaning - Checking Missing Values

```python
df.isnull().sum()
```

### Handling Missing Values

```python
# Replace missing numerical values with mean and categorical values with mode
```

```python
df['temperature'] = df['temperature'].fillna(df['temperature'].mean())
```

```python
df['windspeed'] = df['windspeed'].fillna(df['windspeed'].mean())
```

```python
df['event'] = df['event'].fillna(df['event'].mode() [0])
```

```python
df.isnull().sum()
```

## 4. Date Conversion

```python
# Convert date column into datetime format for time-based analysis
```

```python
df['day']
```

```python
df['day'] = pd.to_datetime(df['day'])
```

```python
df['day']
```

## 5. Feature Engineering

```python
# Create new features from existing date and temperature data
```

```python
df['year'] = df['day'].dt.year
df['month'] = df['day'].dt.month
df['day_name'] = df['day'].dt.day_name()
```

```python
df.head()
```

```python
df['temperature_category'] = df['temperature'].apply(
    lambda x: 'Cold' if x < 15
    else 'Hot' if x > 30
    else 'Normal'
)
df['temperature_category']
```

## 6. Standardizing Column Names

```python
# Standardize column names for better readability
```

```python
df.columns
```

```python
df.columns = df.columns.str.title()
```

```python
df.columns
```

## 7. Statistical Analysis

```python
# Perform descriptive statistics and correlation analysis
```

```python
df.describe()
```

```python
avgTemp = df['Temperature'].mean()
minTemp = df['Temperature'].min()
maxTemp = df['Temperature'].max()
```

```python
print(f"Average Temperature is: {avgTemp}")
print(f"Maximum Temperature is: {maxTemp}")
print(f"Minimum Temperature is: {minTemp}")
```

```python
df.corr(numeric_only=True)
```

## 8. Data Visualization

```python
# Visualize temperature changes over time
```

```python
plt.figure(figsize=(10,5))

plt.plot(
    df['Day'],
    df['Temperature']
)

plt.xlabel("Date")
plt.ylabel("Temperature")

plt.title("Temperature Trend")

plt.show()
```

### Wind Speed and Temperature Relationship

```python
# Analyze relationship between wind speed and temperature
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df['Windspeed'],
    df['Temperature']
)

plt.xlabel("Wind Speed")
plt.ylabel("Temperature")

plt.title("Humidity/Wind Speed vs Temperature")

plt.show()
```

### Weather Condition Distribution

```python
# Visualize frequency of different weather conditions
```

```python
df['Event'].value_counts()
```

```python
df['Event'].value_counts().plot(kind='bar')

plt.title("Weather Conditions")

plt.show()
```

### Temperature Category Analysis

```python
# Visualize created temperature categories
```

```python
df['Temperature_Category'].value_counts().plot(kind='bar')

plt.title("Temperature Categories")

plt.show()
```
